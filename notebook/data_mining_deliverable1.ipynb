```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/workspaces/ProjectDeliverable_1/data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
df.isnull().sum()
```

Missing values were handled using forward fill method to maintain data consistency.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

Duplicate rows were removed to ensure data integrity.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

The TotalCharges column was converted to numeric format for proper analysis.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title("Customer Tenure Distribution")
plt.show()
```

Most customers have tenure between 1-20 months indicating a large number of new customers.

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

The majority of customers did not churn, but a significant percentage left the service.

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

Correlation analysis shows relationships between numerical variables such as tenure and monthly charges.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['MonthlyCharges'])
plt.title("Monthly Charges Outliers")
plt.show()
```

Boxplots help identify extreme values in billing amounts.

Key Insights from Exploratory Data Analysis:

1. Customer tenure shows that many customers are relatively new.
2. Monthly charges vary widely indicating different subscription plans.
3. Some features show moderate correlations which may be useful for predictive modeling.
4. Outliers were detected in monthly charges which may affect regression models.
5. Churn distribution suggests an imbalanced dataset which must be handled during classification modeling.
